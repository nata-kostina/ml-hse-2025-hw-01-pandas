#### 2. [0.5 балла] Есть ли в данных пропуски? В каких колонках? Сколько их в каждой из этих колонок?

### Визуализации и matplotlib

При работе с данными часто неудобно делать какие-то выводы, если смотреть на таблицу и числа в частности, поэтому важно уметь визуализировать данные. Здесь будут описаны ключевые правила оформления графиков для **всех** домашних заданий.

У matplotlib, конечно же, есть [документация](https://matplotlib.org/users/index.html) с большим количеством [примеров](https://matplotlib.org/examples/), но для начала достаточно знать про несколько основных типов графиков:
- plot — обычный поточечный график, которым можно изображать кривые или отдельные точки;
- hist — гистограмма, показывающая распределение некоторой величины;
- scatter — график, показывающий взаимосвязь двух величин;
- bar — столбцовый график, показывающий взаимосвязь количественной величины от категориальной.

В этом задании вы попробуете построить один из них. Не забывайте про базовые принципы построения приличных графиков:
- оси должны быть подписаны, причём не слишком мелко;
- у графика должно быть название;
- если изображено несколько графиков, то необходима поясняющая легенда;
- все линии на графиках должны быть чётко видны (нет похожих цветов или цветов, сливающихся с фоном);
- если отображена величина, имеющая очевидный диапазон значений (например, проценты могут быть от 0 до 100), то желательно масштабировать ось на весь диапазон значений (исключением является случай, когда вам необходимо показать малое отличие, которое незаметно в таких масштабах);
- сетка на графике помогает оценить значения в точках на глаз, это обычно полезно, поэтому лучше ее отрисовывать;
- если распределение на гистограмме имеет тяжёлые хвосты, лучше использовать логарифмическую шкалу.

Еще одна билиотека для визуализации: [seaborn](https://jakevdp.github.io/PythonDataScienceHandbook/04.14-visualization-with-seaborn.html) (обычно сокращается до sns). Это настройка над matplotlib, иногда удобнее и красивее делать визуализации через неё.

**5 пунктов после которых ваши графики не будут прежними:**
- для красивой картинки <code>%config InlineBackend.figure_format = 'retina'</code>
- задаем размер (почти) всех графиков (можно выбрать другие значения) <code>plt.rcParams['figure.figsize'] = 8, 5</code>
- размер шрифта подписей графиков <code>plt.rcParams['font.size'] = 12</code>
- формат в котором сохраняется изображение <code>mpl.rcParams['savefig.format'] = 'pdf'</code>
- sns – seaborn, добавляет решетку <code>sns.set_style('darkgrid')</code>

## 📌 **Алгоритм распределения студентов по курсам**:
1. По умолчанию на каждой дисциплине по выбору у 3 и 4 курсов может учиться 1 группа (до 30 студентов). Исключения описаны ниже. На blended-дисциплинах для четверокурсников количество мест не ограничено.
2. Проводится первая волна отбора.
```python
def arange_courses_wave_1(df, semester):
    #==================================================================
    # Первая волна
    #==================================================================
    
    # Удаляем дублирующиеся курсы
    df.loc[df[f'{semester}_1'] == df[f'{semester}_2'], f'{semester}_2'] = np.nan
    df.loc[df[f'{semester}_1'] == df[f'{semester}_3'], f'{semester}_3'] = np.nan
    df.loc[df[f'{semester}_2'] == df[f'{semester}_3'], f'{semester}_3'] = np.nan
    
    # Формируем две заявки для студентов, выбирающих два предмета
    df_group_21_mop = df[(df[f'nb_{semester}'] == 2) & (df[f'{semester}_1'] != df[f'{semester}_2'])].copy()

    # Переносим второй по выбору предмет в колонку с первым предметов по выдору, чтобы дальше было удобнее работать
    df_group_21_mop.loc[:, f'{semester}_1'] = df_group_21_mop[f'{semester}_2']

    # Конкатенируем с основными данными
    df = pd.concat([df, df_group_21_mop], ignore_index=True)

    # Инициализируем колонки course1 и course2
    df[['course1', 'course2']] = '???'
    df.loc[df[f'nb_{semester}'] == 1, 'course2'] = '-'

    # Курс, на который претендует студент
    df.loc[:, 'candidate_course'] = df[f'{semester}_1']

    # Добавляем информацию о количестве вакантных мест
    df['places'] = df['candidate_course'].map(places).fillna(default_places)

    # Ранжируем студентов для каждого предмета
    df['rank'] = df.groupby('candidate_course')['percentile'].rank(method='first', ascending=True)

    # Студенты, которые выбирают 1 предмет и с высоким перцентилем проходят на предмет первого приоритета
    mask = (df[f'nb_{semester}'] == 1) & (df['rank'] <= df['places'])
    df.loc[mask, 'course1'] = df.loc[mask, 'candidate_course']

    # Студенты, которые выбирают 2 предмета и с высоким перцентилем могут быть приняты на два курса
    df['passed'] = (df['rank'] <= df['places']).astype(int)
    df.loc[df[f'nb_{semester}'] == 2, 'times_accepted'] = 0
    df.loc[df[f'nb_{semester}'] == 2, 'times_accepted'] = df[df[f'nb_{semester}'] == 2].groupby('id')['passed'].transform('sum')

    # Студенты, которые выбирают 2 предмета, но прошедшие отбор только по одному предмету принимаются на этот предмет
    mask = (df[f'nb_{semester}'] == 2) & (df['rank'] <= df['places']) & (df['times_accepted'] == 1)
    df.loc[mask, 'course1'] = df.loc[mask, 'candidate_course']

    # Студенты, которые выбирают 2 предмета и прошедшие отбор по двум предметам принимаются на оба предмета
    mask = (df[f'nb_{semester}'] == 2) & (df['rank'] <= df['places']) & (df['times_accepted'] == 2)
    df.loc[mask, 'row_in_group'] = df.loc[mask].groupby('id').cumcount() # будем заполнять только первую запись в группе, вторую потом удалим
    
    mask = (df[f'nb_{semester}'] == 2) & (df['rank'] <= df['places']) & (df['times_accepted'] == 2) & (df['row_in_group'] == 0)
    df.loc[mask, 'course1'] = df.loc[mask, 'candidate_course']
    df.loc[mask, 'course2'] = df.loc[mask, f'{semester}_2']

    # Удаляем дублированные записи для студентов, которые выбирают 2 предмета
    # Для студентов, выбирающих 2 предмета и не прошедших отбор ни на один курс, удаляем вторую дублирующую записьdf['to_delete'] = False
    df['to_delete'] = False
    df['dup_index'] = df.groupby('id').cumcount()

    mask1 = (df[f'nb_{semester}'] == 2) & (df['times_accepted'] == 0) & (df['dup_index'] > 0)
    df.loc[mask1, 'to_delete'] = True

    # Для студентов, выбирающих 2 предмета и прошедших отбор только на один курс либо на оба курса, удаляем запись с course1=='???'
    mask2 = (df[f'nb_{semester}'] == 2) & (df['times_accepted'] != 0) & (df['course1'] == '???')
    df.loc[mask2, 'to_delete'] = True

    # Удаляем помеченные к удалению данные
    df.drop(df[df['to_delete']].index, inplace=True)
    
    # Удаляем ненужные колонки
    df.drop(['candidate_course', 'places', 'times_accepted', 'rank', 'row_in_group','to_delete','dup_index', 'passed'], axis=1, inplace=True)

    return df

df = arange_courses_wave_1(df, 'fall')
```

```python
def arange_courses_wave_2(df, semester):
    #==================================================================
    # Вторая волна
    #==================================================================
    # Копируем данные в новый датафрейм
    wave_2 = df.copy()
    
    # Определяем курс, на который претендует студент во второй волне
    # 1) Для студентов, которые выбирают 1 курс, но не получили курс в 1-ой волне
    mask1 = (wave_2[f'nb_{semester}'] == 1) & (wave_2['course1'] == '???') & (wave_2[f'{semester}_2'].notna())
    wave_2.loc[mask1, 'candidate_course'] = wave_2.loc[mask1, f'{semester}_2']
    
    # 2) Для студентов, которые выбирают 2 осенних курса и которые не получили второй курс в 1-ой волне
    mask2 = (wave_2[f'nb_{semester}'] == 2) & (wave_2['course1'] != '???') & (wave_2['course2'] == '???') & (wave_2[f'{semester}_3'].notna())
    wave_2.loc[mask2, 'candidate_course'] = wave_2.loc[mask2, f'{semester}_3']
    
    # 3) Для студентов, которые выбирают 2 осенних курса и которые не получили первый курс в 1-ой волне
    mask3 = (wave_2[f'nb_{semester}'] == 2) & (wave_2['course1'] == '???') & (wave_2[f'{semester}_3'].notna())
    wave_2.loc[mask3, 'candidate_course'] = wave_2.loc[mask3, f'{semester}_3']
    
    participants = mask1 | mask3 | mask2
    
    # Считаем количество оставшихся вакантных мест
    nb_accepted_course1 = wave_2.loc[wave_2['course1'] != '???'].groupby('course1').size()
    nb_accepted_course2 = wave_2.loc[(wave_2['course2'] != '???') & (wave_2['course2'] != '-')].groupby('course2').size()
    nb_accepted_total = nb_accepted_course1 + nb_accepted_course2
    
    available_places = places.subtract(nb_accepted_total, fill_value=default_places).clip(lower=0)
    wave_2.loc[participants, 'places'] =wave_2.loc[participants, 'candidate_course'].map(available_places)
    
    # Ранжируем студентов для кажого предмета по перцентилю
    wave_2.loc[participants, 'rank'] = wave_2.loc[participants].groupby('candidate_course')['percentile'].rank(method='first', ascending=True)
    
    # Присваиваем курс студентам, прошедшим отбор
    mask4 = wave_2['rank'] <= wave_2['places']
    # 1) Для студентов, которые выбирают 1 осенний курс, но не получили курс в 1-ой волне
    wave_2.loc[mask1 & mask4, 'course1'] = wave_2['candidate_course']
    
    # 2) Для студентов, которые выбирают 2 осенних курса и которые не получили второй курс в 1-ой волне
    wave_2.loc[mask2 & mask4, 'course2'] = wave_2['candidate_course']
    
    # 3) Для студентов, которые выбирают 2 осенних курса и которые не получили первый курс в 1-ой волне
    wave_2.loc[mask3 & mask4, 'course1'] = wave_2['candidate_course']
    
    # Удаляем ненужные колонки
    wave_2.drop(['candidate_course', 'places', 'rank'], axis=1, inplace=True)
    
    return wave_2

df = arange_courses_wave_2(df, 'fall')    
```

##### Третья волна

```python
def arange_courses_wave_3(df, semester):
    #==================================================================
    # Третья волна
    #==================================================================
    # Копируем данные в новый датафрейм
    wave_3 = df.copy()
    
    # Определяем курс, на который претендует студент в третьей волне
    # Для студентов, которые выбирают 1 осенний курс, но не получили курс в 1-ой и 2-ой волне
    participants = (wave_3[f'nb_{semester}'] == 1) & (wave_3['course1'] == '???') & (wave_3[f'{semester}_3'].notna())
    wave_3.loc[participants, 'candidate_course'] = wave_3.loc[participants, f'{semester}_3']
    
    # Считаем количество оставшихся вакантных мест
    nb_accepted_course1 = wave_3.loc[wave_3['course1'] != '???'].groupby('course1').size()
    nb_accepted_course2 = wave_3.loc[(wave_3['course2'] != '???') & (wave_3['course2'] != '-')].groupby('course2').size()
    nb_accepted_total = nb_accepted_course1 + nb_accepted_course2
    
    available_places = places.subtract(nb_accepted_total, fill_value=default_places).clip(lower=0)
    wave_3.loc[participants, 'places'] = wave_3.loc[participants, 'candidate_course'].map(available_places)
    
    # Ранжируем студентов для кажого предмета по перцентилю
    wave_3.loc[participants, 'rank'] = wave_3.loc[participants].groupby('candidate_course')['percentile'].rank(method='first', ascending=True)
    
    # Присваиваем курс студентам, прошедшим отбор
    mask_rank = wave_3['rank'] <= wave_3['places']
    # 1) Для студентов, которые выбирают 1 осенний курс, но не получили курс в 1-ой и 2-ой волне
    wave_3.loc[participants & mask_rank, 'course1'] = wave_3['candidate_course']
    
    # Удаляем ненужные колонки
    wave_3.drop(['candidate_course', 'places', 'rank'], axis=1, inplace=True)
    
    return wave_3

df = arange_courses_wave_3(df, 'fall')
```
